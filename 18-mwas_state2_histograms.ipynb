```r
file <- files[i]

library(data.table)

file_in <- fread(file)

head(file_in)

library(data.table)
library(ggplot2)

# Set the number of rows to sample
x <- 5000

# Randomly sample x rows from the data.table
sampled_data <- file_in[sample(.N, x)]

# Plot a histogram of the p values
ggplot(sampled_data, aes(x = p)) +
  geom_histogram(binwidth = 0.01, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = paste0("Histogram of p values\n", basename(unique(file_in$stats))), x = "p values", y = "Frequency") +
  theme_minimal()
```

```r
library(data.table)
library(ggplot2)

# Initialize a list to store the plots
plot_list <- list()

for (file in files) {
    file_in <- fread(file)
    print

    head(file_in)
    
    # Set the number of rows to sample
    x <- 5000
    
    # Randomly sample x rows from the data.table
    sampled_data <- file_in[sample(.N, x)]
    
    # Plot a histogram of the p values
    p <- ggplot(sampled_data, aes(x = p)) +
      geom_histogram(binwidth = 0.01, fill = "blue", color = "black", alpha = 0.7) +
      labs(title = paste0("Histogram of p values\n", basename(unique(file_in$stats))), x = "p values", y = "Frequency") +
      theme_minimal()
    
    # Store the plot in the list
    plot_list[[file]] <- p
    
    # Save the plot to a file
    ggsave(filename = paste0("histogram_", basename(file), ".png"), plot = p)
}

# Optionally, return or print the list of plots
plot_list
```
