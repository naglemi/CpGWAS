```python
import os
import pandas as pd
from rpy2.robjects import r, pandas2ri
from rpy2.robjects.packages import importr

pandas2ri.activate()
base = importr('base')
readRDS = base.readRDS

df = pd.read_csv("12-OUT_matched_SNP_meth_cov_outputs.csv")
traits = ["bp", "mdd", "scz"]

for trait in traits:
    df['stage2_path'] = df['path'].apply(lambda x: x.replace('.rds', f'_gwas_stat_{trait}_results.rds'))
    output_file = f"16a3-OUT_stage2_MWAS_{trait}.csv"
    header_written = False
    
    for i in range(len(df)):
        stage2_in = readRDS(df.loc[i, 'stage2_path'])
        stage1_in = readRDS(df.loc[i, 'path'])

        stage1_models = list(stage1_in.slots['models'])
        stage2_models = list(stage2_in.slots['MWASmodels'])

        if len(stage1_models) != len(stage2_models):
            raise ValueError("Files don't match")

        data_list = []
        for j in range(len(stage1_models)):
            model1 = stage1_models[j]
            model2 = stage2_models[j]

            this_z = model2.rx2('z')[0]
            this_p = model2.rx2('p')[0]
            this_n = model2.rx2('n')[0]
            this_pos = model1.rx2('methylationPosition')[0]
            this_stats = stage2_in.rx2('summary_stats_path')[0]
            this_scaff = stage1_in.rx2('scaffoldIdentifier')[0]

            data_list.append({
                'z': this_z,
                'p': this_p,
                'n': this_n,
                'pos': this_pos,
                'stats': this_stats,
                'scaff': this_scaff
            })

        combined_df = pd.DataFrame(data_list)
        combined_df.to_csv(output_file, index=False, header=not header_written, mode='a')
        header_written = True
```
