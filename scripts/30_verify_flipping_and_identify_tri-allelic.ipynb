```r
library(data.table)
library(dplyr)

# List of chromosomes
chromosomes <- 1:22  # Assuming chromosomes 1 through 22

# Base paths (adjust as necessary)
path_libd <- "/expanse/lustre/projects/jhu152/naglemi/mwas/gwas_flipped/"
path_ref <- "/expanse/lustre/projects/jhu152/naglemi/mwas/gwas_flipped/"

# Loop through each chromosome
for (chr in chromosomes) {
  # Construct file names
  file_libd <- paste0(path_libd, "libd_chr", chr, ".pvar")
  file_ref <- paste0(path_ref, "ref_EUR_flipped_chr", chr, ".pvar")

  # Read files
  dt_libd <- fread(file_libd)[, .(CHROM = `#CHROM`, POS, REF, ALT)]
  dt_ref <- fread(file_ref)[, .(CHROM = `#CHROM`, POS, REF, ALT)]

  # Find non-matching rows
  non_matching_libd <- anti_join(dt_libd, dt_ref, by = c("CHROM", "POS", "REF", "ALT"))
  non_matching_ref <- anti_join(dt_ref, dt_libd, by = c("CHROM", "POS", "REF", "ALT"))

  # Report non-matching details
  if (nrow(non_matching_libd) > 0 || nrow(non_matching_ref) > 0) {
    cat(sprintf("%d SNPs on chromosome %d don't match, as shown below:\n", nrow(non_matching_libd) + nrow(non_matching_ref), chr))
    
    if (nrow(non_matching_libd) > 0) {
      print("Rows in libd not in ref:")
      print(non_matching_libd)
    }
    
    if (nrow(non_matching_ref) > 0) {
      print("Rows in ref not in libd:")
      print(non_matching_ref)
    }
  } else {
    cat(sprintf("All SNPs on chromosome %d match perfectly.\n\n", chr))
  }
    cat("\n")
}
```

```r
library(data.table)
library(dplyr)

# Prepare to collect all non-matching entries
all_non_matching <- list()

# List of chromosomes
chromosomes <- 1:22  # Assuming chromosomes 1 through 22

# Base paths (adjust as necessary)
path_libd <- "/expanse/lustre/projects/jhu152/naglemi/mwas/gwas_flipped/"
path_ref <- "/expanse/lustre/projects/jhu152/naglemi/mwas/gwas_flipped/"

# Loop through each chromosome
for (chr in chromosomes) {
  # Construct file names
  file_libd <- paste0(path_libd, "libd_chr", chr, ".pvar")
  file_ref <- paste0(path_ref, "ref_EUR_flipped_chr", chr, ".pvar")

  # Read files
  dt_libd <- fread(file_libd)[, .(CHROM = `#CHROM`, POS, REF, ALT)]
  dt_ref <- fread(file_ref)[, .(CHROM = `#CHROM`, POS, ID, REF, ALT)]

  # Find non-matching rows
  non_matching_libd <- anti_join(dt_libd, dt_ref, by = c("CHROM", "POS", "REF", "ALT"))
  non_matching_ref <- anti_join(dt_ref, dt_libd, by = c("CHROM", "POS", "REF", "ALT"))

  # Accumulate non-matching entries, including ID from ref
  if (nrow(non_matching_ref) > 0) {
    all_non_matching[[chr]] <- non_matching_ref[, .(CHROM, POS, ID)]
  }

  # Report non-matching details
  if (nrow(non_matching_libd) > 0 || nrow(non_matching_ref) > 0) {
    cat(sprintf("%d SNPs on chromosome %d don't match, as shown below:\n", nrow(non_matching_libd) + nrow(non_matching_ref), chr))
    
    if (nrow(non_matching_libd) > 0) {
      print("Rows in libd not in ref:")
      print(non_matching_libd[, .(CHROM, POS, REF, ALT)])
    }
    
    if (nrow(non_matching_ref) > 0) {
      print("Rows in ref not in libd:")
      print(non_matching_ref[, .(CHROM, POS, ID, REF, ALT)])
    }
  } else {
    cat(sprintf("All SNPs on chromosome %d match perfectly.\n\n", chr))
  }
    cat("\n")
}

# Combine all non-matching entries into a single dataframe
final_non_matching <- rbindlist(all_non_matching)
final_non_matching <- unique(final_non_matching)
# Save the final list to CSV
fwrite(final_non_matching, "30-OUT_multi-allele_exclude_list.csv", col.names = TRUE)
```

```r
# Second loop to verify IDs against CHROM and POS
for (chr in chromosomes) {
  # Read the reference file again
  dt_ref <- fread(paste0(path_ref, "ref_EUR_flipped_chr", chr, ".pvar"))[, .(CHROM = `#CHROM`, POS, ID)]

  # Subset final_non_matching for the current chromosome
  verify_ids <- final_non_matching[CHROM == chr]

  # Check ID correspondence
  verification_result <- merge(verify_ids, dt_ref, by = c("CHROM", "POS", "ID"), all.x = TRUE, all.y = FALSE)
  missing_ids <- verification_result[is.na(ID)]

  if (nrow(missing_ids) > 0) {
    cat(sprintf("Verification failed for %d entries on chromosome %d as shown below:\n", nrow(missing_ids), chr))
    print(missing_ids)
  } else {
    cat(sprintf("All IDs on chromosome %d have been verified correctly.\n\n", chr))
  }
}
```
