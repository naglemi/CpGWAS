## Run over all

```python
import os
import pandas as pd
import glob

# Get current working directory
cwd = os.getcwd()

# Set working directory to the parent directory
os.chdir("..")

# Read all CSV files
file_list = glob.glob("**/*-DNAm_stats.csv", recursive=True)

print(len(file_list))

# Open a file in write mode to write the header
with open("15a2-OUT_DNAm_vs_cor_intermediate.csv", "w") as f:
    header_written = False
    
    for i, file in enumerate(file_list, start=1):
        if i % 20 == 0:
            print(i)
        temp_data = pd.read_csv(file)
        
        # Write the header only once
        if not header_written:
            temp_data.to_csv(f, index=False)
            header_written = True
        else:
            temp_data.to_csv(f, index=False, header=False, mode='a')
```
