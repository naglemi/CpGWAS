```bash
# Path to the PLINK binary
plink_path="./plink2"

# Loop through chromosomes 1 to 22
for chr in {1..22}
do
  echo "Processing chromosome $chr..."

  # Define file paths for this chromosome
  input_pfile="ref_EUR_allele-match-libd_chr${chr}"
  output_pfile="ref_EUR_allele-match-a2-libd_chr${chr}"

  # Enclose paths in quotes to handle special characters and spaces
  exclude_list_path="/expanse/lustre/projects/jhu152/naglemi/mwas/CpGWAS/scripts/30-OUT_multi-allele_exclude_list_pt2-from32a2.txt"

  # Check if the exclusion list file is not empty
  if [ -s "$exclude_list_path" ]; then
    # Run PLINK to exclude the SNPs listed in the exclusion list
    $plink_path --pfile "$input_pfile" --exclude "$exclude_list_path" --make-pgen --out "$output_pfile"

    echo "PLINK has been executed to exclude SNPs for chromosome $chr."
  else
    echo "No SNPs to exclude for chromosome $chr, or an error occurred in finding the exclusion list."
  fi

  # Add multiple blank lines for readability between iterations
  echo -e "\n\n\n\n\n"
done
```
